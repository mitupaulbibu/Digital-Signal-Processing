```python
def sort(x):
    for i in range(len(x)):
        for j in range(len(x) - 1):
            if x[j]>x[j+1]:
                temp = x[j]
                x[j] = x[j+1]
                x[j + 1] = temp
    return x
    
x = [34, 0, -2, 99, 10, 23, 1]
y= sort(x)
print(y)
```

```python
#1. generate sinusoid
import numpy as np
import matplotlib.pyplot as plt
F = 10
Fs = 400
L = 50
x = np.zeros(L)
for n in range(L):
    x[n] = np.sin(2*np.pi*F*n/Fs) 
#print(x)
plt.plot(x)
plt.stem(x)
```

```python
#unit impulse
d = np.zeros(10)
d[0] = 1
#plt.plot(d)
plt.stem(d)
```

```python
#unit impulse
#d = np.zeros(10)
#d[0] = 1
n = range(-10, 10)
d = np.zeros(len(n))
d[len(n)//2] = 1
#plt.plot(d)
plt.stem(n,d)
```

```python
#unit step
n = range(-10, 11)
d = np.zeros(len(n))
for i in range(len(n)//2):
    d[len(n)//2+i] = 1
plt.stem(n,d)
```

```python
#ramp
n = range(-10, 11)
d = np.zeros(len(n))
for i in range(len(n)//2):
    d[len(n)//2+i] = 1 + i
plt.stem(n,d)
```
